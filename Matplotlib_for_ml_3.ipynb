```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2, 2, 1) #Can also write plt.subplot(221)
plt.subplot(2, 2, 2) #Can also write plt.subplot(222)
plt.subplot(2, 2, 3) #Can also write plt.subplot(223)
plt.subplot(2, 2, 4) #Can also write plt.subplot(224)

#plt.subplot(2, 2, 5) This will give error, 2*2 = 4 graphs can be made only

plt.show()
```

```python
plt.figure(figsize = (10, 5))

plt.subplot(221)
plt.subplot(222)
plt.subplot(223)
plt.subplot(224)

plt.show()
```

```python
plt.figure(figsize = (10, 6))

plt.subplot(2,2,1)
plt.pie([1])

plt.subplot(2,2,2)
plt.pie([1,2])

plt.subplot(2,2,3)
plt.plot([1,2,3,4,5,6],[1,2,3,4,5,6])

plt.subplot(2,2,4, projection = 'polar', facecolor = 'k')

plt.show()
```

```python
plt.subplot(2, 2, 1) 
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3) 
plt.subplot(2, 2, 4) 

plt.show()
```

```python
plt.subplot(2, 2, 1) 
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3) 
plt.subplot(2, 2, 4)

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 1) 
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3) 
plt.subplot(2, 2, 4) 

plt.subplots_adjust(left=0.1, bottom=0.1, right=2, top=2, wspace=1, hspace=1)

plt.show()
```

### To save the graph as image file in the system

```python
plt.subplot(2, 2, 1) 
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3) 
plt.subplot(2, 2, 4)

plt.savefig("grpimg")
plt.show()
```

```python
plt.subplot(2, 2, 1) 
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3) 
plt.subplot(2, 2, 4)

plt.savefig("grpimg", dpi = 100, facecolor = 'k')
plt.show()
```
